```julia
using PyPlot
include("Taylor.jl")
using ADT
using ODE
```

```julia
function ec1(t, x̄)
    
    dx = x̄[2]
    dy = (α + β * cos(t)) * sin(x̄[1])
    
    return [dx; dy]
end
```

```julia
x̄ = [0.0, 1.0]
t0 = 0.0
tf = 50 * pi

α = 0.0
β = 1.0

t = linspace(t0, tf, 100000);
```

```julia
(t1, sol1) = ode45(ec1, x̄, t);
```

```julia
x1 = map(y -> y[1], sol1)
y1 = map(y -> y[2], sol1);
```

```julia
plot(x1, y1)
```

```julia
(t2, sol2) = integradorT2(ec1, t0, tf, x̄, 100000, 1e-60);
```

```julia
x2 = map(y -> y[1], sol2)
y2 = map(y -> y[2], sol2);
```

```julia
plot(x2, y2)
# xlim(-0.8, -0.7)
# ylim(-0.25, 0.25)
```
